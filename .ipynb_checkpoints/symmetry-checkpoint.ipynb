# Molekülgeometrie und Symmetry

## Molekülgeometrie

### a) kartesische Koordinaten
Das planare Ameisensäure-Molekül ist gegeben als:
- $r(C-O_1) = 1.202 ~ \overset{\circ}{A}$
- $r(C-O_2) = 1.343 ~ \overset{\circ}{A}$
- $r(O_2-H_1) = 0.972 ~ \overset{\circ}{A}$
- $r(C-H_2) = 1.097 ~ \overset{\circ}{A}$
- $\alpha(H_2-C-O_1) = 235.1 ^\circ$
- $\beta(O_1-C-O_2) = 124.9 ^\circ$
- $\gamma(C-O_2-H_1) = 106.3 ^\circ$

Wir legen das C-Atom in den Koordinatenursprung und $O_1$ auf die x-Achse:
$$
\vec{v}(C) = \begin{pmatrix} 0 \\ 0 \end{pmatrix}~~~
\vec{v}(O_1) = \begin{pmatrix} 1.202 \\ 0 \end{pmatrix}
$$

Der Winkel des $r(C-O_1)$-Vektors zum $r(C-O_2)$-Vektor ist gegeben als 124.9$^\circ$. Wir benutzen die zwei-dimensionale Rotationsmatrix:

$$
\hat{R}(\theta) = 
\begin{bmatrix}
\cos(\theta) & -\sin(\theta) \\
\sin(\theta) & \cos(\theta) 
\end{bmatrix}
$$

Und können dadurch die Position von $O_2$ bestimmen:
$$
\vec{v}(O_2) = \underbrace {\hat{R}\left( 124.9^\circ \right)}_{\text{Rotation}}~~
\underbrace{\vec{v}(O_1)}_{\text{Vektor}} ~~
\underbrace{\cdot \frac{r(C-O_1)}{r(C-O_2)}}_{\text{Normierung}}
$$

$$
\vec{v}(O_2) = 
\begin{bmatrix}
\cos(2.180 \pi) & -\sin(2.180 \pi) \\
\sin(2.180 \pi) & \cos(2.180 \pi) 
\end{bmatrix}
\begin{pmatrix} 1.202 \\ 0 \end{pmatrix}
\frac{1.202}{1.343} = \begin{pmatrix} -0.768 \\ 1.102 \end{pmatrix}
$$ 

Als weiteres Beispiel folgt:
$$
\vec{v}(H_1) = \hat{R}\left( \alpha(H_2-C-O_1)\right)~~
\left(\vec{v}(C) - \vec{v}(O_2)\right) ~~
\cdot \frac{r(O_2-H_1)}{r(C-O_2)} ~~
\underbrace{+ \vec{v}(O_2)}_{\text{Verschiebung}}
$$

$$
\vec{v}(H_1) = 
\begin{bmatrix}
\cos(1.855 \pi) & -\sin(1.855 \pi) \\
\sin(1.855 \pi) & \cos(1.855 \pi) 
\end{bmatrix}
\begin{pmatrix} 0.768 \\ -1.102 \end{pmatrix}
\frac{1.343}{0.972} + \begin{pmatrix} -0.768 \\ 1.102 \end{pmatrix}
= \begin{pmatrix} -0.159 \\ 1.859 \end{pmatrix}
$$

Insgesamt ergeben sich die Koordinaten dadurch als:

Atom | x | y
--- | --- | ---
C     | 0.0     | 0.0   
O1    | 1.202   | 0.0   
O2    | -0.768  | 1.102 
H1    | -0.159  | 1.859 
H2    | -0.628  | -0.899

### c) Interne Koordinaten: die Z-Matrix

Die Z-Matrix ist eine spezielle Art, Verbindungsinformationen in einem Koordinatensystem darzustellen. Diese hat folgende Struktur:

 Index | Symbol |      Index      |   $r$           | Index           | $\alpha$             | Index           | $\theta$
 ---   |    --- |    ---     | ---           |   ---      | ---            | ---        | --- 
 Index | Symbol |            |               |           |                |            |
 Index | Symbol | Referenz 1 | Bindungslänge |           |                |            |
 Index | Symbol | Referenz 1 | Bindungslänge | Referenz 2 | Bindungswinkel |            |
 Index | Symbol | Referenz 1 | Bindungslänge | Referenz 2 | Bindungswinkel | Referenz 3 | Diederwinkel
 Index | Symbol | Referenz 1 | Bindungslänge | Referenz 2 | Bindungswinkel | Referenz 3 | Diederwinkel
 Index | Symbol | Referenz 1 | Bindungslänge | Referenz 2 | Bindungswinkel | Referenz 3 | Diederwinkel

 
Das bedeutet, für jedes Atom in der Z-Matrix braucht man 8 Datenpunkte, um dessen Position im Raum eindeutig zu beschreiben. Die Gesamtgröße der Z-Matrix ist somit $8n-6$, es handelt sich also nicht um eine sehr kompakte Datenstruktur (Vergleich: kartesische Koordinaten $3n$).
 
Anders als in kartesischen Koordinaten kann man aus einer Z-Matrix aber direkt Bindungsinformationen ablesen und manipulieren. Deswegen sind Z-Matrices (oder Varianten davon) zum Beispiel in kristallographischen Strukturlöseprogrammen, wie ShelXL, durchaus üblich. 

Für **1,2,5-oxadiazole** ergibt sich folgende Z-Matrix:

 Index | Symbol |      Index      |   $r$           | Index           | $\alpha$             | Index           | $\theta$
 ---   |    --- |    ---     | ---           |   ---      | ---            | ---        | --- 
 1 | O |            |               |           |                |            |
 2 | N |  1 | 138 pm  |           |                |            |
 3 | C |  2 | 130 pm | 1 | 105.8° |            |
 4 | C |  3 | 142 pm | 2 | 109.0° | 1 | 0°
 5 | N |  4 | 130 pm | 3 | 109.0° | 2 | 0°
 6 | H |  3 | 108 pm | 4 | 142.1° | 5 | 180°
 7 | H |  4 | 108 pm | 3 | 142.1° | 6 | 180°
 
 Wie Sie sehen können, ist der Aufbau der Z-Matrix willkürlich, solange jede Atomposition eindeutig definiert ist. Man kann zum Beispiel keinen Bezug auf eine Atomposition nehmen, die vorher noch nicht definiert wurde.


## Dipolmoment

Das Dipolmoment in der Punktladungsnäherung ist definiert als:
$$
\vec{\mu} = \sum_i^N q_i \vec{r}_i
$$

Für Fluormethan ergibt sich:
$$
\begin{aligned}
\vec{\mu} &= -0.053  \cdot \begin{pmatrix} 0 \\ 0 \\  0 \end{pmatrix}
-0.276  \cdot \begin{pmatrix} 0 \\ 0 \\ 1.403 \end{pmatrix} \\
&~~~~+0.110  \cdot \begin{pmatrix} 1.054 \\ 0 \\ -0.360 \end{pmatrix}
+0.110  \cdot \begin{pmatrix} -0.527 \\ -0.913 \\ -0.360 \end{pmatrix}
+0.110  \cdot \begin{pmatrix} -0.527 \\ 0.913 \\ -0.360 \end{pmatrix}\\
&= \begin{pmatrix} 0.0 \\ 0.0 \\ -0.51 \end{pmatrix} ~~[e/\overset{\circ}{A}]\\
|\vec{\mu}| &= 0.51~[e/\overset{\circ}{A}] = 2.43~Debye
\end{aligned}
$$

Für Pyrimidin ergibt sich:
$$
\vec{\mu} = \begin{pmatrix} -0.28 \\ 0.0 \\ -0.17 \end{pmatrix}  [e/\overset{\circ}{A}]\\
|\vec{\mu}| = 0.33~[e/\overset{\circ}{A}] = 1.56~Debye
$$


## Molekülsymmetry

### a) Gruppenmultiplikationstabellen

Symmetryoperationen im Wassermolekül:
<img src="Pictures/watersymm.png" alt="Drawing" style="width: 200px;"/>


 - | E | C$_2$ | $\sigma_v^1$ | $\sigma_v^2$
--- | --- | --- | --- | ---
E   | E   | C$_2$ | $\sigma_v^1$ | $\sigma_v^2$
C$_2$ | C$_2$ | E | $\sigma_v^2$ | $\sigma_v^1$
$\sigma_v^1$ | $\sigma_v^1$ | $\sigma_v^2$ | E | C$_2$
$\sigma_v^2$ | $\sigma_v^2$ | $\sigma_v^1$ | C$_2$ | E


$$
C_2 \cdot \sigma_v^1 =
\begin{bmatrix}
-1 & 0 & 0 \\
0  & -1 & 0 \\
0  & 0 & 1 
\end{bmatrix}
\begin{bmatrix}
-1 & 0 & 0 \\
0  & 1 & 0 \\
0  & 0 & 1 
\end{bmatrix}
=
\begin{bmatrix}
1 & 0 & 0 \\
0  & -1 & 0 \\
0  & 0 & 1 
\end{bmatrix}
=
\sigma_v^2
$$

Zwei senkrechte Spiegelebenen erzeugen eine zweizählige Drehachse:
$$
\sigma_v^1 \cdot \sigma_v^2 =
\begin{bmatrix}
-1 & 0 & 0 \\
0  & 1 & 0 \\
0  & 0 & 1 
\end{bmatrix}
\begin{bmatrix}
1 & 0 & 0 \\
0  & -1 & 0 \\
0  & 0 & 1 
\end{bmatrix}
=
\begin{bmatrix}
-1 & 0 & 0 \\
0  & -1 & 0 \\
0  & 0 & 1 
\end{bmatrix}
=
C_2
$$

Weil es nur eine C$_n$ mit $n \geq 2$ gibt, keine weiteren senkrechten Drehachsen und keine horizontale Spiegelebene $\sigma_h$, aber vertikale Spiegelebenen $\sigma_v$, ist die Punktgruppe C$_{2v}$.

### c) Vergleich von Punktgruppen
$C_ {\infty v}$ (zum Beispiel HCN) vs $D_{\infty h}$ (zum Beispiel CO$_2$): Letzteres hat eine horizontale Spiegelebene. Weil $C_{2n} ~\in C_\infty$, weil also jede unendliche Drehachse auch geradzahlige Drehachsen enthält, entsteht ein Inversionszentrum.

### d) Vergleich von Punktgruppen
Beispiel für $C_{3v}$: NH$_3$

Beispiel für $D_{3h}$: BF$_3$

BF$_3$ ist planar und hat daher die zusätzliche $\sigma_h$ Spiegelebene.

### e) Isomere

1,1-Dichlorethan: $C_{2v}$
$E$-1,2-Dichlorethen: $C_{2v}$
$Z$-1,2-Dichlorethen: $C_{2h}$

### f) Noch mehr Übung
Ameisensäure: $C_s$
1,2,5-oxadiazole: $C_{2v}$

### g) Dipolmoment und Punktgruppe
Eine Punktgruppe mit Inversionszentrum kann kein Dipolmoment besitzen.

Ein Dipolmoment horizontal zu einer Spiegelebene wird aufgebhoben.

Diese Kriterien sind notwendig, aber nicht hinreichend. Die Moleküle ortho-Dichlorobenzen und meta-Dichlorobenzen zum Beispiel können zwar ein Dipolmoment besitzen, dieses ist aber praktisch nicht vorhanden.
