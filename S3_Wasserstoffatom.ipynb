# Seminar 3: Das Wasserstoffatom

## Lösen der Schrödinger-Gleichung

Die Schrödinger-Gleichung für das Wasserstoffatom lautet:

$$
\left( \underbrace{\frac{-1}{2} \Delta}_{=\hat{T}_{kin}}  - \underbrace{\frac{1}{| \mathbf{\hat{r}}|}}_{=\hat{V}_{Coulomb}} \right) \Psi(\mathbf{r}) = E\Psi(\mathbf{r})
$$

Hier ist $\mathbf{r} = x \hat{i} + y \hat{j} + z \hat{k} = \left(x~y~z\right)^T$ ein Vektor in kartesischen Koordinaten. Da es sich aber um ein zentrosymmetrisches Problem handelt, bietet sich die Transformation zu Kugelkoordinaten an:

$$
\begin{aligned}
x &\rightarrow |r| \cdot \sin(\theta) \cos(\varphi)  ~~ &r \in \left[0, \infty\right] \\
y &\rightarrow |r| \cdot \sin(\theta) \sin(\varphi)  ~~ &\theta \in \left[0, \pi\right] \\
z &\rightarrow |r| \cdot \cos(\theta)             ~~ &\varphi \in \left[0, 2\pi\right]
\end{aligned}
$$

Das bedeutet, unsere Wellenfunktion transformiert nach $\Psi(\mathbf{r}) \rightarrow \Psi(r, \theta, \varphi)$. Der Coulomb-Operator $\hat{V}$ ist winkelunabhängig und transformiert einfach zu $1/r$. Der Laplace-Operator für die kinetische Energie sieht in Kugelkoordinaten allerdings anders aus:

$$
\Delta = \frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} \right)
+ \frac{1}{r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial}{\partial \theta} \right) 
+ \frac{1}{r^2 \sin^2(\theta)} \frac{\partial^2 }{\partial \varphi^2}
$$

Dieser Operator suggeriert bereits, dass wir die radiale und winkelabhängige Bewegung separieren können:

$$
\Psi(r, \theta, \varphi) = R(r) \cdot Y(\theta, \varphi)
$$

Setzen wir dies in die SG ein, folgt:

$$
\left( \frac{-1}{2r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} \right)
+ \frac{-1}{2r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial}{\partial \theta} \right) 
+ \frac{-1}{2r^2 \sin^2(\theta)} \frac{\partial^2 }{\partial \varphi^2} 
+ \hat{V} \right) R(r) \cdot Y(\theta, \varphi) = E R(r) \cdot Y(\theta, \varphi)
$$

Das sieht viel komplizierter aus, als es eigentlich ist. Die winkelabhängigen Operatoren kommutieren mit $R(r)$, die winkelunabhängigen Operatoren kommutieren mit $Y(\theta, \varphi)$:

$$
\begin{aligned}
&Y \cdot \frac{-1}{2r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial R}{\partial r} \right) \\
&+ R \cdot \frac{-1}{2r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial Y}{\partial \theta} \right) \\
&+ R \cdot \frac{-1}{2r^2 \sin^2(\theta)} \frac{\partial^2 Y}{\partial \varphi^2} \\
&+ Y \cdot \hat{V} R  \\
&= R Y \cdot E 
\end{aligned}
$$

Wir multiplizieren mit $- 2\cdot r^2$ und dividieren durch $RY$:

$$
\begin{aligned}
\underbrace{
R^{-1}\left[ \frac{\partial}{\partial r} \left(r^2 \frac{\partial R}{\partial r} \right) 
-2\cdot r^2 \left( \hat{V} R - E \right) \right]
}_{\text{radialabhängig}}
+ \underbrace{
Y^{-1} \left[ \cdot \frac{1}{\sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial Y}{\partial \theta} \right) 
+\cdot \frac{1}{\sin^2(\theta)} \frac{\partial^2 Y}{\partial \varphi^2}\right]
}_{\text{winkelabhängig}} = 0
\end{aligned}
$$



## Wie sehen die Lösungen aus?

Beide Seiten müssen gleich groß und konstant sein. Die winkelabhängige Gleichung ist dabei eine spezielle Differentialgleichung: Die Laplace-Gleichung aus der klassischen Elektrodynamik. Deren Lösungen sind die sogenannten Legendre-Funktionen (Kugelflächenfunktionen, "spherical harmonics"). Mit dieser Information lassen sich auch die Radialfunktionen exakt bestimmen.

Die Kugelflächenfunktionen haben folgende Form:

$$
Y_{l, m}(\theta, \varphi) = \underbrace{\epsilon \sqrt{\frac{(2l+1)}{4\pi} \frac{(l - |m|)!}{(l + |m|)!}} e^{im\varphi}}_{\text{Normierung}}~\cdot~
\underbrace{P_l^m(cos(\theta))}_{\text{Legendre-Funktionen}}
~~\text{mit}~\epsilon=\begin{cases}(-1)^m~\forall~m\geq 0 \\ 1~\forall~m \leq 0 \end{cases}
$$

Die Lösungen der Radialfunktion haben folgende Form ($a_0$ ist der Bohr-Radius):

$$
R_{n,l}(r) = \underbrace{\sqrt{\frac{2}{na_0}^3 \frac{(n-l-1)!}{2n[(n+l)!]^3}}}_{\text{Normierung}}~\cdot~
\underbrace{\left(\frac{2r}{na_0}\right)^l \left[L_{n-l-1}^{2l+1}({2r}/{na_0})\right]}_{\text{Laguerre-Polynome}}~\cdot~
\underbrace{e^{-r/na_0}}_{\text{exponentieller Abfall}}
$$

Mit folgenden Quantenzahlen:

$$
\begin{aligned}
n &= 1,2,3... & \rightarrow \text{Hauptquantenzahl} \\
l &= 0,1,2... <n & \rightarrow \text{Nebenquantenzahl} \\
m &= -l, (-l+1), ..., (l-1), +l & \rightarrow \text{Magnetquantenzahl} \\
\end{aligned}
$$

Die Energieniveaus hängen nur von der Hauptquantenzahl $n$ ab:

$$
E_{n} = \left[ \frac{-m_e e^4}{8 \varepsilon_0^2 \hbar^2} \right] \frac{1}{n^2}
$$

Die Lösungen der Schrödinger-Gleichung für das Wasserstoffatom sind somit nicht besonders schön. Stark vereinfacht können wir sie uns folgendermaßen vorstellen:

$$
\Psi_{n, l, m} = \text{Abstandsabhängigkeit} \times \text{Raumausrichtung} \times \text{Quantisierung}
$$

Als praktisches Beispiel:
Die Kugelflächenfunktion für $l=1$ und $m_l=0$ ($p$-Orbital) sieht so aus:

$$
Y_{l=1,m=0} = \sqrt{\frac{3}{4\pi}} \cos(\theta)
$$

Die Radialfunktion für das dazugehörige $n=2$ und $l=1$ sieht so aus:

$$
R_{n=2, l=1} = \frac{1}{\sqrt{24}} a_0^{-3/2} \frac{r}{a} e^{-r/2a_0}
$$

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import sph_harm, genlaguerre, factorial
from matplotlib import cm, colors
plt.xkcd()
plt.style.use("seaborn-ticks")
plt.rcParams["legend.handlelength"] = 0.8
plt.rcParams["legend.framealpha"] = 0.8
font_name = "Arial"
font_size = 12
plt.rcParams.update({"font.sans-serif": font_name, "font.size": font_size})

def Rad(r):
    return 1/np.sqrt(24) * 1**(-3/2) * r * np.exp(-r/2)

def Ylm(theta):
    return  np.cos(theta) * np.sqrt(3/(4*np.pi))

thets = np.linspace(0, 2, 100) * np.pi
rs = np.linspace(0, 20, 100)
r = thets

fig = plt.figure(figsize=(15, 5))
ax = plt.subplot(1, 3, 1)
ax.plot(rs, Rad(rs))
ax.title.set_text("Radial part")
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel("r")
ax.set_ylabel("R(r)")

ax = plt.subplot(1, 3, 2, projection='polar')
ax.plot(thets+(r<0)*np.pi, np.abs(Ylm(thets)))
ax.set_rticks([])  # less radial ticks
ax.set_rlabel_position(-22.5)  # get radial labels away from plotted line
ax.grid(True)
ax.title.set_text("Angular part")

#plt.show()


Z   = 1  # Anzahl der Elementarladungen
a_0 = 1. # Bohrscher Atomradius

def R(n,l, r):
    """Radialfunktion"""
    Z_n = 2*Z/(n*a_0)
    root = np.sqrt(Z_n**3 * factorial(n-l-1)/(2*n*factorial(n+l)))
    rho = Z_n * r
    L = np.polyval(genlaguerre(n-l-1, 2*l+1), rho)
    return root * np.exp(-0.5*rho) * rho**l * L

def psi(n,l,m, r,theta,phi):
    """de.wikipedia.org/wiki/Wasserstoffatom#Mathematische_Details"""
    R_nl = R(n,l, r)
    Y_lm = sph_harm(m,l,phi,theta) 
    return R_nl * Y_lm

def plot(n,l,m, x,z, r,theta,phi, ax):
    ax.title.set_text(r"$R(r) \cdot Y(\theta, \varphi )$")
    s_nlm = np.abs(psi(n,l,m, r,theta,phi))
    ax.pcolormesh(x, z, s_nlm, cmap=plt.get_cmap("seismic_r"), 
                  norm = colors.Normalize(vmin=-0.05,vmax=0.05))
    plt.gca().set_aspect("equal") 
    ax.set_xlim(x.min(),x.max())
    ax.set_ylim(z.min(),z.max())
    
ax = plt.subplot(1, 3, 3)
ax.set_xticks([])
ax.set_yticks([])
n_max = 2
for n in range(2,n_max+1):
    for l in [1]:
        plot_range = (4*n + 4*l) * a_0
        x_1d = np.linspace(-plot_range,plot_range,521)
        y = 0
        z_1d = np.linspace(-plot_range,plot_range,521)
        x,z = np.meshgrid(x_1d,z_1d)
        r     = np.sqrt(x**2 + y**2 + z**2)
        theta = np.arctan2(np.sqrt(x**2+y**2), z )
        phi   = np.arctan2(y, x)
        for m in range(1, l+1):
            plot(n,l,m, x,z, r,theta,phi,ax)

plt.show()
```

### b) Lyman-Serie
Die Lyman-Serie ist eine spezielle spektroskopische Serie des Wasserstoffatoms, wo Übergänge von angeregten Zuständen $n'$ zum Grundzustand $n=1$ beobachtet werden.

$$
\frac{1}{\lambda} = R_\text{H} \left( \frac{1}{n^2} - \frac{1}{n'^2} \right) 
$$

Mit der Rydberg-Konstante $R_\text{H} \approx ~1.0968\cdot 10^7~\text{m}^{-1}~\approx~\frac{13.6~\text{eV}}{hc}$ und $\lambda = 9.38\cdot 10^{-8}~m$.

Diese Gleichung müssen Sie nach $n'$ umstellen, um einen genäherten Wert zu erhalten.

### a) Zeichnen der Radialteile

Folgende Orbitalwellenfunktionen sind gegeben:

$$
\begin{aligned}
1s && \psi_{100} = N_{100} e^{-r/a_0} \\
2s && \psi_{200} = N_{200} e^{-r/2a_0} a_0^{-3/2} \left(2 - \frac{r}{a_0} \right) \\
2p_x && \psi_{211}=\sqrt{\frac{3}{4 \pi}} \frac{1}{2 \sqrt{6}} a_{0}^{-3 / 2} \frac{r}{a_{0}} e^{-r / 2 a_{0}} \sin(\theta) \cos(\varphi) \\
2p_y && \psi_{21-1}=\sqrt{\frac{3}{4 \pi}} \frac{1}{2 \sqrt{6}} a_{0}^{-3 / 2} \frac{r}{a_{0}} e^{-r / 2 a_{0}} \sin(\theta) \cos(\varphi)\\
2p_z && \psi_{210}=N_{210} \frac{r}{a_{0}} e^{-r / 2 a_{0}} \cos(\theta)
\end{aligned}
$$


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
plt.xkcd()
plt.style.use("seaborn-ticks")
plt.rcParams["legend.handlelength"] = 0.8
plt.rcParams["legend.framealpha"] = 0.8
font_name = "Arial"
font_size = 12
plt.rcParams.update({"font.sans-serif": font_name, "font.size": font_size})

def psi100(r):
    return 1/np.sqrt(np.pi) * 1**(-3/2) * np.exp(-3*r/2)

def psi200(r):
    return 1/np.sqrt(4*np.pi) * 1/np.sqrt(2*np.sqrt(2)) * 1**(-3/2) * np.exp(-r/2)*(2-r)

def psi211(r):
    return np.sqrt(3/(4*np.pi)) * (1/(2*np.sqrt(6))) * 1**(-3/2) * r * np.exp(-r/2)

rs = np.linspace(0, 10, 1000)

fig, ax = plt.subplots(1,3, figsize=(15,5), sharey=True, sharex=True)
ax[0].set_yticks([])
ax[0].set_xticks([])
ax[0].set_xlabel("r")
ax[0].set_ylabel("R(r)")
ax[0].axhline(y=0, xmin=0, xmax=20, color="k", linestyle="--", alpha=0.5)
ax[0].axhline(y=0, xmin=0, xmax=20, color="k", linestyle="--", alpha=0.5)
ax[0].plot(rs, psi100(rs), color="black")
ax[0].title.set_text("$\psi_{100}$")
ax[0].fill_between(rs, psi100(rs), color="crimson", alpha=0.5)

ax[1].axhline(y=0, xmin=0, xmax=20, color="k", linestyle="--", alpha=0.5)
ax[1].plot(rs, psi200(rs), color="black")
ax[1].title.set_text("$\psi_{200}$")
ax[1].fill_between(rs, (psi200(rs) > 0)* psi200(rs), color="crimson", alpha=0.5)
ax[1].fill_between(rs, (psi200(rs) < 0)* psi200(rs), color="royalblue", alpha=0.5)

ax[2].axhline(y=0, xmin=0, xmax=20, color="k", linestyle="--", alpha=0.5)
ax[2].plot(rs, psi211(rs), color="black")
ax[2].title.set_text("$\psi_{211}$")
ax[2].fill_between(rs, (psi211(rs) > 0)* psi211(rs), color="crimson", alpha=0.5)
ax[2].fill_between(rs, (psi211(rs) < 0)* psi211(rs), color="royalblue", alpha=0.5)

plt.show()
```


### d) Normierungsbedingung

$$
\begin{aligned}
\text{allgemein:  }& 1 = |A|^2 \int^\infty \Psi_j^* \Psi_i d\tau \\
\text{kartesisch:  }& 1 = |A|^2 \int^\infty_{-\infty}\int^\infty_{-\infty}\int^\infty_{-\infty} \Psi_j^*(x, y, z) \Psi_i(x,y,z) dx dy dz \\
\text{Kugelkoordinaten:   } & 1 = |A|^2 \int_0^{2\pi} \int_0^{\pi} \int_0^\infty \Psi_j^*(r, \theta, \varphi) \Psi_i(r, \theta, \varphi) r^2 \sin(\theta) dr d\theta d\varphi
\end{aligned}
$$

- Sie müssen in der Lage sein, die Atomorbitale zu normieren (Bestimmung von A).
- Sie sollten die Orthogonalität der Radialfunktionen und der Kugelflächenfunktionen beweisen können.
- Sie sollten einfache Erwartungswerte berechnen können.


## Erwartungswerte

### Der Erwartungswert der Wahrscheinlichkeit

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\begin{aligned}
\braket{A}_a^b = \braket{\Psi_j | \Psi_i}_a^b = \int_a^b \Psi_j \Psi_i d\tau
\end{aligned}
$$

### Der Erwartungswert der Energie

Am Beispiel des 1s-Orbitals des Wasserstoffatoms in Kugelkoordinaten:

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\begin{aligned}
\braket{E_{100}} &= \braket{\psi_{100}(r, \theta, \varphi) \left| \hat{H} \right| \psi_{100}(r, \theta, \varphi)} \\
&= \braket{\psi_{100}(r, \theta, \varphi) | \frac{-\Delta}{2} - \frac{1}{|\hat{r}|} | \psi_{100}(r, \theta, \varphi)} \\
&= \int \psi_{100}(r, \theta, \varphi) \frac{-1}{2} \left( 
\frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} \right)
+ \frac{1}{r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial}{\partial \theta} \right) 
+ \frac{1}{r^2 \sin^2(\theta)} \frac{\partial^2 }{\partial \varphi^2} 
- \frac{1}{|\hat{r}|} \right) 
\psi_{100}(r, \theta, \varphi) d\tau
\end{aligned}
$$



### Der Erwartungswert des Abstands

Der Ortsoperator der Wellenfunktion agiert folgendermaßen:

$$
\hat{x}_i \Psi(x_i) = x_i \cdot \Psi(x_i)
$$

Wir können also einen Abstandsoperator als den Betrag des Ortsoperators definieren und in Kugelkoordinaten darstellen:

$$
\begin{aligned}
|\hat{x}_i| \Psi(x_i) &= |x_i| \cdot \Psi(x_i) \\
\hat{r} \Psi(r, \theta, \phi) &= |\mathbf{r}| \cdot \Psi(r, \theta, \phi) \\
&= \sqrt{
\left[|r| \sin(\theta)\cos(\phi)\right]^2 
+ \left[|r| \sin(\theta)\sin(\phi)\right]^2 
+ \left[|r| \cos(\theta)\right]^2} \cdot \Psi(r, \theta, \phi) \\
&= \sqrt{|r|^2 \cdot \underbrace{
\left[ 
\sin(\theta)^2 \cdot 
\left( \cos(\phi)^2 + \sin(\phi)^2 \right) 
+ \cos(\theta)^2
\right]}_{=1}} \cdot \Psi(r, \theta, \phi) \\
&= r \cdot \Psi(r, \theta, \phi)
\end{aligned}
$$

Der Erwartungswert des Abstands in Kugelkoordinaten ist also:

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\braket{r} = \braket{\Psi|\hat{r}|\Psi} = \braket{\Psi|r|\Psi} = \int r \cdot |\Psi|^2 d\tau
$$

Für das 1s-Orbital des Wasserstoffatoms erhält man:

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\begin{aligned}
\braket{r} &= \frac{4}{a_0^3} \int_0^\infty r^3 e^{-2r/a_0} dr \\
&= \frac{4}{a_0^3} \cdot \frac{6}{\left( \frac{2}{a_0} \right)^4} =\frac{3}{2} a_0 \\
&\approx 1.5 ~ \text{Angström}
\end{aligned}
$$

Als Vergleich, der van-der-Waals-Radius von Wasserstoff liegt zwischen 1.2 bis 2 Angström.
