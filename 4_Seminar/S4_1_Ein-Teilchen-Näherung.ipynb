## Wasserstoffähnliche Atome - die "Independent Particle Approximation"

Im vorherigen Seminar haben wir das Problem des Wasserstoffatoms gelöst, wo ein Elektron ($1~e^{-}$) einen Kern mit der Kernladungszahl $Z=1$ umkreist.

Jetzt wollen wir das Problem für $n~e^{-}$ lösen, also für beliebig viele Elektronen, die um einen Kern beliebiger Ladung kreisen.

Die Schrödinger-Gleichung dafür lautet:

$$
\begin{aligned}
\hat{H} &\Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right) = E \Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right) & \text{mit } \vec{r}_i = \left( x_i, y_i, z_i\right)^T \\
\hat{H} &= \hat{T}_{\text{kin}} + \hat{V}_{\text{Coulomb}} 
\end{aligned}
$$

Schauen wir uns erst die kinetische Energie an. Diese beinhaltet den Laplace-Operator $\Delta$, also die zweite Ableitung der Wellenfunktion nach den Komponenten $x, y, z$. Für eine Funktion mehrerer solcher Komponenten ist die kinetische Energie gegeben als das **Totale Differential**:

$$
\begin{aligned}
\hat{T}_{\text{kin}} = \frac{-\hbar^2}{2 m_e} \Delta &= \frac{-\hbar^2}{2 m_e} \sum_i^N \Delta_i \\
&= \frac{-\hbar^2}{2 m_e} \left( \frac{\partial^2}{\partial {x}_1^2} + \frac{\partial^2}{\partial {y}_1^2} + 
\frac{\partial^2}{\partial {z}_1^2} + ... + \frac{\partial^2}{\partial {x}_N^2} + \frac{\partial^2}{\partial {y}_N^2} + \frac{\partial^2}{\partial z_N^2}    \right)
\end{aligned}
$$

Schauen wir uns jetzt die potentielle Energie an. Das Potential zwischen zwei geladenen Teilchen wird durch das **Coulomb-Potential** beschrieben. In der **Punktladungsnäherung** ist das Potential vieler geladener Teilchen also die Summe von vielen Coulomb-Potentialen:

$$
\begin{aligned}
\hat{V}_{\text{Coulomb}} &= \underbrace{\sum_i^N \frac{-Z}{|\vec{r}_i|}}_{\text{Kern-Elektron-Anziehung}} 
+ \frac{1}{2}\underbrace{\sum_{i \neq j}^N \frac{1}{|\vec{r}_i - \vec{r}_j|}}_{\text{Elektron-Elektron-Abstoßung}} \\
\\
&= \sum_i^N \hat{V}_{i}^{\text{eK}} + \frac{1}{2} \sum_{i \neq j}^N \hat{V}_{ij}^{\text{ee}}
\end{aligned}
$$

In allen bisherigen Beispielen haben wir gesehen, dass wir den Hamilton-Operator in Komponenten separieren konnten, die jeweils nur von einer Koordinate abhingen. In diesen Fällen lässt sich die Lösung der Differentialgleichung als Produkt der Einzellösungen darstellen.

**Hier können wir das nicht, weil der Operator $\hat{V}_{ij}^{\text{ee}}$ auf zwei Komponenten gleichzeitig wirkt.** Durch die Elektron-Elektron-Wechselwirkung ist die Wellenfunktion nicht mehr separierbar.

### Ein-Teilchen-Näherung (Independent Particle Approximation)

Aber wie sähe die Lösung aus, wenn es die Elektron-Elektron-Wechselwirkung einfach nicht gäbe? Wir ignorieren als Ansatz $\hat{V}_{ij}^{\text{ee}}$. Dann hängt der Hamilton-Operator nur noch von den Koordinaten einzelner Elektronen ab. Das heißt, wir können $\Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right)$ separieren:

$$
\begin{aligned}
\hat{H} &\Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right) = \left( \frac{-\hbar^2}{2 m_e} \sum_i^N \Delta_i + \sum_i^N \hat{V}_{i}^{\text{eK}} \right) \Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right) \\
& \Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right)  \rightarrow \psi_{1}(\vec{r}_1) \cdot \psi_{2}(\vec{r}_2) \cdot ... \cdot \psi_{N}(\vec{r}_N) = \prod_i^N \psi_{i}(\vec{r}_i)
\end{aligned}
$$

Zur Vereinfachung der Indices lassen wir die Koordinatenabhängigkeit implizit fallen. Schauen wir uns exemplarisch nur eine Komponente des Hamilton-Operators an, der auf diese Wellenfunktion wirkt:

$$
\begin{aligned}
\left(\frac{-\hbar^2}{2 m_e} \Delta_1 + \hat{V}_{1}^{\text{eK}}\right) \prod_i^N \psi_{i}
= \prod_{i=2}^N \psi_{i} \cdot \underbrace{\left(\frac{-\hbar^2}{2 m_e} \Delta_1 + \hat{V}_{1}^{\text{eK}}\right) \psi_{1}}_{=\hat{h}_1 \psi_{1} = \varepsilon_i \psi_1}
\end{aligned}
$$

Alle Komponenten, auf die der Operator nicht wirkt, kommutieren. Die Lösung der Gleichung $\hat{h}_1 \psi_{1}$ kennen wir - diese ist das Schrödinger-Problem von einem Elektron im Feld eines Kerns. Die Lösungen für $\psi_1$ sind also die Wasserstoff-Atomorbitale mit der Ein-Elektron-Energie $\varepsilon_i$, die lediglich durch die Kernladung $Z$ skaliert werden. Es lässt sich leicht zeigen, dass die Gesamtenergie die Summe der Ein-Elektron-Energien $E = \sum_i^N \varepsilon_i$ ist.

## Slater-Regeln

Wir benötigen sinnvolle, physikalisch motivierte Näherungen, um die Elektron-Elektron-Wechselwirkung in einer Ein-Teilchen-Näherung darzustellen. Eine stark vereinfachte Näherung, die sehr einfache Berechnungen erlaubt, wurde von Slater vorgestellt.

$$
\begin{aligned}
\text{Ein-Elektron-Energie:    }  & ~~ \varepsilon_{nl} = - \frac{1}{2} \frac{(Z-\sigma)^2}{n'^2} \\
\text{Gesamtenergie:    }  & ~~ E = \sum_{i=1}\varepsilon_{nl}^{(i)} \\
\text{Effektive Kernladungszahl:    }  & ~~ Z_{eff} = Z- \sigma \\
\end{aligned}
$$


### Elektron 4. Periode im K-Atom: 

$$
\begin{aligned}
\text{Elektronenkonfiguration } & ~~1s^2 2s^2 2p^6 3s^2 3p^6 4s^1 \\ \\
\text{Abschirmungskonstante } & ~~\sigma = \underbrace{2 \cdot 1.0}_{1s} + \underbrace{2 \cdot 1.0}_{2s} + \underbrace{6 \cdot 1.0}_{2p} + \underbrace{2 \cdot 0.85}_{3s} + \underbrace{6 \cdot 0.85}_{3p} = 16.80 \\
\text{effektive Kernladung } & ~~Z_{eff} = 19 - 16.80 = 2.20
\end{aligned}
$$

#### Daraus können wir die Energie des höchsten besetzten Niveaus bestimmen, die der Ionisierungsenergie nach Koopman entspricht:

$$
\begin{aligned}
\text{Ein-Elektron-Energie:  } & ~~ \varepsilon_{n=4, l=0} = - \frac{1}{2} \frac{2.20^2}{3.7^2} = -0.1768~E_H = -4.81~eV \\
\text{Koopman's Theorem:  } & ~~ IE = (-1) \cdot \varepsilon_{HOMO} =  - \varepsilon_{4,0} = 4.81~eV
\end{aligned}
$$

#### Wir können außerdem die Gesamtenergie bestimmen und daraus die Ionisierungsenergie nach der Delta-Methode:

$$
\begin{aligned}
\text{K-Atom:  } & ~~ \sum_{i=1}\varepsilon_{nl}^{(i)}
= \underbrace{2 \varepsilon_{1,0}+ 2 \varepsilon_{2,0} + 6 \varepsilon_{2,1} + 2 \varepsilon_{3,0} + 6 \varepsilon_{3,1}}_{1s^2 2s^2 2p^6 3s^2 3p^6} + \underbrace{\varepsilon_{4,0}}_{4s^1}\\ \\
\text{K$^+$-Ion:  } & ~~ \sum_{i=1}\varepsilon_{nl}^{(i)}
= \underbrace{2 \varepsilon_{1,0}+ 2 \varepsilon_{2,0} + 6 \varepsilon_{2,1} + 2 \varepsilon_{3,0} + 6 \varepsilon_{3,1}}_{1s^2 2s^2 2p^6 3s^2 3p^6}\\ \\
\Delta E &= E_{K^+} - E_{K} = IE = \varepsilon_{4,0} = 4.81 eV \\ \\
\text{Experiment:  } IE &= 4.34~eV
\end{aligned}
$$

### Elektron 3. Periode im Ar-Atom:

$$
\begin{aligned}
\text{Elektronenkonfiguration } & ~~1s^2 2s^2 2p^6 3s^2 3p^6 \\ \\
\text{Abschirmung } & ~~\sigma = \underbrace{2 \cdot 1.0}_{1s} + \underbrace{2 \cdot 0.85}_{2s} + \underbrace{6 \cdot 0.85}_{2p} + \underbrace{2 \cdot 0.35}_{3s} + \underbrace{5 \cdot 0.35}_{3p} = 11.25 \\
\text{effektive Kernladung } & ~~Z_{eff} = 18 - 16.80 = 6.75
\end{aligned}
$$

#### Ionisierungsenergie nach Koopman:

$$
\begin{aligned}
\text{Ein-Elektron-Energie:  } & ~~ \epsilon_{n=3, l=1} = - \frac{1}{2} \frac{6.75^2}{3^2} = -2.5313~E_H = -68.88~eV \\
\text{Koopman's Theorem:  } & ~~ IE = (-1) \cdot \epsilon_{HOMO} =  - \epsilon_{3,1} = 68.88~eV
\end{aligned}
$$

#### Wir können außerdem die Gesamtenergie bestimmen und daraus die Ionisierungsenergie nach der Delta-Methode:

$$
\begin{aligned}
\text{Ar-Atom:  } & ~~ \sum_{i=1}\epsilon_{nl}^{(i)}
= \underbrace{2 \epsilon_{1,0}+ 2 \epsilon_{2,0} + 6 \epsilon_{2,1} + 2 \epsilon_{3,0} + 6 \epsilon_{3,1}}_{1s^2 2s^2 2p^6 3s^2 3p^6}\\ \\
\text{Ar$^+$-Ion:  } & ~~ \sum_{i=1}\epsilon_{nl}^{(i)}
= \underbrace{2 \epsilon_{1,0}+ 2 \epsilon_{2,0} + 6 \epsilon_{2,1} + 2 \epsilon_{3,0}}_{1s^2 2s^2 2p^6 3s^2} + 5 \epsilon_{3,1}
\end{aligned}
$$

Hier müssen wir jetzt beachten, dass sich die Abschirmung in der dritten Schale durch das Entfernen eines Elektrons geändert hat:

$$
\begin{aligned}
\text{Elektronenkonfiguration Ar$^+$ } & ~~1s^2 2s^2 2p^6 3s^2 3p^5 \\ \\
\text{Abschirmung Ar$^+$ } & ~~\sigma = \underbrace{2 \cdot 1.0}_{1s} + \underbrace{2 \cdot 0.85}_{2s} + \underbrace{6 \cdot 0.85}_{2p} + \underbrace{2 \cdot 0.35}_{3s} + \underbrace{4 \cdot 0.35}_{3p} = 10.95 \\
\text{Ein-Elektron-Energie Ar$^+$-Ion:  } & ~~ = \epsilon_{n=3, l=1} = - \frac{1}{2} \frac{\left(18 - 10.95\right)^2}{3^2} = -2.8006~E_H = -76.21~eV
\end{aligned}
$$

Daraus ergibt sich die Ionisierungsenergie:

$$
\begin{aligned}
IE &= \Delta E = E_{Ar^+} - E_{Ar}  \\
&=  \epsilon_{3, 1}(Ar^+) - \epsilon_{3, 1}(Ar) \\
&= 7 \cdot (-76.21~eV) - 8 \cdot (-68.88~eV) \\
&= 17.57~eV \\
\text{Experiment:  } IE &= 15.76~eV
\end{aligned}
$$

## Die Säkulargleichungen

### Die Bedeutung des Produktansatzes

Stellen wir uns vor, es gibt einen **effektiven Hamilton-Operator**, der die Elektron-Elektron-Wechselwirkung mittelt und dadurch separierbar wird:

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\hat{H} \approx \hat{H}^{eff} = \sum_i^N \hat{T}^{kin}_i + \sum_i^N \hat{V}^{eK}_i + \sum_i^N \braket{\hat{V}^{ee}_i}
$$

Dann können wir eine **Produktwellenfunktion** finden, die diese Differentialgleichung löst:

$$
\Psi = \prod_i^N \psi_i(x_i)
$$

Diese Produktwellenfunktion gewährleistet aber nicht die **Ununterscheidbarkeit von Elektronen.** Jedes Elektron $i$ ist in diesem Ansatz unabhängig von jedem anderen Elektron $j$ und hängt nur von den eigenen Koordinaten $x_i, y_i, z_i$ ab. Dadurch gewährleistet dieser Ansatz auch nicht die Einhaltung des **Pauli-Prinzips**. Um diese physikalischen Notwendigkeiten herzustellen, muss das Produkt **antisymmetrisch** sein, was sich in der sogenannten **Slater-Determinante** darstellen lässt.

Wenn man das tut, findet man explizite Ausdrücke für $\braket{\hat{V}^{ee}_i}$ und somit für $\hat{H}^{eff}$. Ein Beispiel für einen solchen effektiven Hamilton-Operator (*mean-field operator*) ist der Fock-Operator aus der Hatree-Fock-Theorie.

**Mit einem effektiven Hamilton-Operator sind alle Elektronen identisch und erfüllen das Pauli-Prinzip.** Das bedeutet, es reicht aus, die Schrödinger-Gleichung für ein einziges effektives Teilchen zu lösen:

$$
\hat{h}^{eff}_i \psi_i = \varepsilon_i \psi_i
$$

Die Lösungen für $\psi_i$ nennt man **Molekülorbitale**. Weil alle Elektronen identisch sind, können wir diese MOs einfach nach dem Aufbauprinzip besetzen.
