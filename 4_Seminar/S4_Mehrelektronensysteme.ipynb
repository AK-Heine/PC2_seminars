## Wasserstoffähnliche Atome - die "Independent Particle Approximation"

Im vorherigen Seminar haben wir das Problem des Wasserstoffatoms gelöst, wo ein Elektron ($1~e^{-}$) einen Kern mit der Kernladungszahl $Z=1$ umkreist.

Jetzt wollen wir das Problem für $n~e^{-}$ lösen, also für beliebig viele Elektronen, die um einen Kern beliebiger Ladung kreisen.

Die Schrödinger-Gleichung dafür lautet:

$$
\begin{aligned}
\hat{H} &\Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right) = E \Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right) & \text{mit } \vec{r}_i = \left( x_i, y_i, z_i\right)^\dagger \\
\hat{H} &= \hat{T}_{\text{kin}} + \hat{V}_{\text{Coulomb}} 
\end{aligned}
$$

Schauen wir uns erst die kinetische Energie an. Diese beinhaltet den Laplace-Operator $\Delta$, also die zweite Ableitung der Wellenfunktion nach den Komponenten $x, y, z$. Für eine Funktion mehrerer solcher Komponenten ist die kinetische Energie gegeben als das **Totale Differential**:

$$
\begin{aligned}
\hat{T}_{\text{kin}} = \frac{-\hbar^2}{2 m_e} \Delta &= \frac{-\hbar^2}{2 m_e} \sum_i^N \Delta_i \\
&= \frac{-\hbar^2}{2 m_e} \left( \frac{\partial^2}{\partial {x}_1^2} + \frac{\partial^2}{\partial {y}_1^2} + 
\frac{\partial^2}{\partial {z}_1^2} + ... + \frac{\partial^2}{\partial {x}_N^2} + \frac{\partial^2}{\partial {y}_N^2} + \frac{\partial^2}{\partial z_N^2}    \right)
\end{aligned}
$$

Schauen wir uns jetzt die potentielle Energie an. Das Potential zwischen zwei geladenen Teilchen wird als das **Coulomb-Potential** beschrieben. Wir können also *annehmen*, dass das Potential vieler geladener Teilchen als Summe von Coulomb-Potentialen dargestellt werden kann.

$$
\begin{aligned}
\hat{V}_{\text{Coulomb}} &= \underbrace{\sum_i^N \frac{-Z}{|\vec{r}_i|}}_{\text{Kern-Elektron-Anziehung}} 
+ \frac{1}{2}\underbrace{\sum_{i \neq j}^N \frac{1}{|\vec{r}_i - \vec{r}_j|}}_{\text{Elektron-Elektron-Abstoßung}} \\
\\
&= \sum_i^N \hat{V}_{i}^{\text{eK}} + \frac{1}{2} \sum_{i \neq j}^N \hat{V}_{ij}^{\text{ee}}
\end{aligned}
$$

In allen bisherigen Beispielen haben wir gesehen, dass wir den Hamilton-Operator in Komponenten separieren konnten, die jeweils nur von einer Koordinate abhingen. In diesen Fällen lässt sich die Lösung der Differentialgleichung als Produkt der Einzellösungen darstellen.

**Hier können wir das nicht, weil der Operator $\hat{V}_{ij}^{\text{ee}}$ auf zwei Komponenten gleichzeitig wirkt.** Durch die Elektron-Elektron-Wechselwirkung ist die Wellenfunktion nicht mehr separierbar.

### Independent Particle Approximation

Aber wie sähe die Lösung aus, wenn es die Elektron-Elektron-Wechselwirkung einfach nicht gäbe? Wir ignorieren als ersten Ansatz $\hat{V}_{ij}^{\text{ee}}$. Dann hängt der Hamilton nur noch von den Koordinaten einzelner Elektronen ab. Das heißt, wir können $\Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right)$ separieren:

$$
\begin{aligned}
\hat{H} &\Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right) = \left( \frac{-\hbar^2}{2 m_e} \sum_i^N \Delta_i + \sum_i^N \hat{V}_{i}^{\text{eK}} \right) \Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right) \\
& \Psi\left(\vec{r}_1, \vec{r}_2, ..., \vec{r}_N\right)  \rightarrow \psi_{1}(\vec{r}_1) \cdot \psi_{2}(\vec{r}_2) \cdot ... \cdot \psi_{N}(\vec{r}_N) = \prod_i^N \psi_{i}(\vec{r}_i)
\end{aligned}
$$

Ab jetzt lassen wir die Koordinatenabhängigkeit implizit fallen. Schauen wir uns exemplarisch nur eine Komponente des Hamilton-Operators an, der auf diese Wellenfunktion wirkt:

$$
\begin{aligned}
\left(\frac{-\hbar^2}{2 m_e} \Delta_1 + \hat{V}_{1}^{\text{eK}}\right) \psi_{1} \psi_{2} ... \psi_{N} 
= \psi_{2} ... \psi_{N} \cdot \underbrace{\left(\frac{-\hbar^2}{2 m_e} \Delta_1 + \hat{V}_{1}^{\text{eK}}\right) \psi_{1}}_{=\hat{h}_1 \psi_{1} = \varepsilon_i \psi_1}
\end{aligned}
$$

Alle Komponenten, auf die der Operator nicht wirkt, kommutieren. Die Lösung der Gleichung $\hat{h}_1 \psi_{1}$ kennen wir - diese ist das Schrödinger-Problem von einem Elektron im Feld eines Kerns. Die Lösungen für $\psi_1$ sind also die Wasserstoff-Atomorbitale mit der Ein-Elektron-Energie $\varepsilon_i$. Wie sich leicht zeigen lässt, ist die Gesamtenergie die Summe der Ein-Elektron-Energien $E = \sum_i^N \varepsilon_i$.

## Slater-Regeln

Wir benötigen sinnvolle, physikalisch motivierte Näherungen, um die Elektron-Elektron-Wechselwirkung in einer Ein-Teilchen-Näherung darzustellen. Eine stark vereinfachte Näherung, die sehr einfache Berechnungen erlaubt, wurde von Slater vorgestellt.

$$
\begin{aligned}
\text{Ein-Elektron-Energie:    }  & ~~ \epsilon_{nl} = - \frac{1}{2} \frac{(Z-\sigma)^2}{n'^2} \\
\text{Gesamtenergie:    }  & ~~ E = \sum_{i=1}\epsilon_{nl}^{(i)} \\
\text{Effektive Kernladungszahl:    }  & ~~ Z_{eff} = Z- \sigma \\
\end{aligned}
$$

## Die Säkulargleichungen

### Die Bedeutung des Produktansatzes

Stellen wir uns vor, es gibt einen **effektiven Hamilton-Operator**, der die Elektron-Elektron-Wechselwirkung mittelt und dadurch separierbar wird:

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\hat{H} \approx \hat{H}^{eff} = \sum_i^N \hat{T}^{kin}_i + \sum_i^N \hat{V}^{eK}_i + \sum_i^N \braket{\hat{V}^{ee}_i}
$$

Dann können wir eine **Produktwellenfunktion** finden, die diese Differentialgleichung löst:

$$
\Psi = \prod_i^N \psi_i(x_i)
$$

Diese Produktwellenfunktion macht aber einen physikalischen Fehler: Sie gewährleistet nicht die **Ununterscheidbarkeit von Elektronen.** Um diese wieder herzustellen, führt man die sogenannte **Slater-Determinante** ein (nächtes Seminar).

Wenn man das tut, findet man explizite Ausdrücke für $\braket{\hat{V}^{ee}_i}$ und somit für $\hat{H}^{eff}$. Ein Beispiel für einen solchen effektiven Hamilton-Operator (*mean-field operator*) ist der Fock-Operator aus der Hatree-Fock-Theorie.

**Mit einem effektiven Hamilton-Operator sind alle Elektronen identisch und erfüllen das Pauli-Prinzip.** Das bedeutet, es reicht aus, die Schrödinger-Gleichung für ein einziges effektives Teilchen zu lösen:

$$
\hat{h}^{eff}_i \psi_i = \varepsilon_i \psi_i
$$

Die Lösungen für $\psi_i$ nennt man **Molekülorbitale**. Weil alle Elektronen identisch sind, können wir diese MOs einfach nach dem Aufbauprinzip besetzen.

---

### Linear Combination of Atomic Orbitals (LCAO)

Es gibt keinen allgemeinen analytischen Ausdruck für die Molekülorbitale $\psi_i$. Um Lösungen zu finden, können wir aber das Superpositionsprinzip ausnutzen: Jede Wellenfunktion lässt sich als Superposition anderer Wellenfunktionen darstellen.

$$
\psi_i = \sum_j c_j \phi_j
$$

Hierbei ist $\phi_j$ eine bekannte Funktion und $c_j$ ein unbekannter Koeffizient. Man bezeichnet diese als **Basis**. Der obige Ausdruck ist exakt, wenn man unendlich viele Basisfunktionen zur Repräsentation von $\psi_i$ heranzieht.

Alle Funktionen $\phi_j$, die dem Schrödinger-Problem genügen, sind als Basisfunktion geeignet (endlich, kontinuierlich, stetig differenzierbar etc.). Die Funktionen, die das System am besten beschreiben, ergeben auch die geringste Energie (**Rayleigh-Ritz Prinzip**). Dadurch können wir einen Algorithmus entwerfen, der mittels **Variation** die besten Koeffizienten $c_j$ findet, die die **Gesamtenergie** minimieren:

$$ 
\begin{aligned}
\sum_j c_j \phi_j  & \xrightarrow[{\text{von } c_j}]{\text{Variation}}  \sum_j c_j' \phi_j \\
\sum_i \varepsilon_i & \xrightarrow[{\text{Gesamtenergie}}]{\text{minimiert}}  \sum_i \varepsilon_i'
\end{aligned}
$$

Wie genau dieser Algorithmus für größere Systeme aussieht, ist Teil des nächsten Seminars (MO-Theorie). Im Folgenden schauen wir uns als Beispiel ein zweiatomiges Molekül mit zwei Elektronen an.

### Das Wasserstoff-Molekül

Das Wasserstoffmolekül besteht aus zwei Wasserstoff-Atomen an den Positionen A und B. Wir wählen von jedem Atom nur ein Atomorbital aus, um die Molekülorbitale zu entwickeln. Wir vernachlässigen den Index $i$ und die komplex-konjugierten Koeffizienten:

$$
\psi = c_A \phi_A + c_B \phi_B
$$

Der Erwartungswert der Gesamtenergie ist:

$$
E = \frac{\int \psi \hat{H} \psi d\tau}{\int \psi \psi d\tau}
$$

Wir setzen ein:

$$
\begin{aligned}
\int \psi \hat{H} \psi d\tau &= \int 
\left(c_A \phi_A + c_B \phi_B \right) \hat{H} \left(c_A \phi_A + c_B \phi_B \right) d\tau \\
&= \int \left(c_A \phi_A + c_B \phi_B \right) \hat{H} c_A \phi_A d\tau 
+ \int \left(c_A \phi_A + c_B \phi_B \right) \hat{H} c_B \phi_B d\tau \\
&= c_A^2 \underbrace{\int \phi_A \hat{H} \phi_A d\tau}_{=\epsilon_{AA} = \alpha}
+ c_A c_B \underbrace{\int \phi_B \hat{H} \phi_A d\tau}_{=\epsilon_{BA} = \beta}
+ c_A c_B \underbrace{\int \phi_A \hat{H} \phi_B d\tau}_{=\epsilon_{AB} = \beta}
+ c_B^2 \underbrace{\int c_B \phi_B \hat{H} \phi_B d\tau}_{=\epsilon_{BB} = \alpha} \\
&= c_A^2 \alpha + c_B^2 \alpha + 2 c_A c_B \beta
\end{aligned}
$$

Hierbei ist $\alpha$ die Energie des Atomorbitals $A$ und $\beta$ die Energie des Überlappungsintegrals $AB$. Da es ein symmetrisches homoatomares System ist, gilt $\beta_{AB}=\beta_{BA}$ und $\alpha_{AA}=\alpha_{BB}$.

Es folgt für den Nenner:

$$
\begin{aligned}
\int \psi \psi d\tau &= \int \left(c_A \phi_A + c_B \phi_B \right)^2 d\tau \\
&= c_A^2 \int \phi_A^2 d\tau + 2 c_A c_B \underbrace{\int \phi_A \phi_B d\tau}_{=S} + c_B^2 \int \phi_B^2 d\tau \\
&= c_A^2 + c_B^2 + 2 c_A c_B S
\end{aligned}
$$

Wir erhalten also für die Energie:

$$
E = \frac{c_A^2 \alpha + c_B^2 \alpha + 2 c_A c_B \beta}{c_A^2 + c_B^2 + 2 c_A c_B S}
$$

#### Anwenden des Variationsprinzips

Wir suchen die Wellenfunktion des Grundzustands, also die Koeffizienten $c_A$ und $c_B$, die die Gesamtenergie $E$ minimieren. Das bedeutet, wir bilden folgendes Differential:

$$
\frac{\partial E}{\partial c_A} = \frac{\partial E}{\partial c_B} = 0
$$

Dafür müssen wir die Ableitung berechnen:

$$
\begin{aligned}
\frac{\partial E}{\partial c_A} &= \frac{\partial}{\partial c_A} 
\frac{\overbrace{c_A^2 \alpha + c_B^2 \alpha + 2 c_A c_B \beta}^{=u}}{\underbrace{c_A^2 + c_B^2 + 2 c_A c_B S}_{=v}} 
= \frac{\frac{\partial u}{\partial c_A}v - u\frac{\partial v}{\partial c_A} }{v^2} \\
&= \frac{\left(2 c_A \alpha + 2 c_B \beta\right)\left(c_A^2 + c_B^2 + 2 c_A c_B S\right) 
- \left(c_A^2 \alpha + c_B^2\alpha + 2 c_A c_B \beta\right)\left(2 c_A + 2 c_B S \right) 
}{\left(c_A^2 + c_B^2 + 2 c_A c_B S \right)^2} \\
&= \frac{\left(2 c_A \alpha + 2 c_B \beta\right)\left(c_A^2 + c_B^2 + 2 c_A c_B S\right)}{\left(c_A^2 + c_B^2 + 2 c_A c_B S \right)^2}
- \frac{\left(c_A^2 + c_B^2 + 2 c_A c_B S\right)\left(2 c_A + 2 c_B S\right)}{\left(c_A^2 + c_B^2 + 2 c_A c_B S \right)^2}
\end{aligned}
$$

Wir setzen die obige Definition der Energie ein und machen Gebrauch davon, dass $\left(c_A^2 + c_B^2 + 2 c_A c_B S \right) = \int \psi \psi d\tau = 1$ für orthonormale Orbitale:

$$
\begin{aligned}
\frac{\partial E}{\partial c_A} &= 
\left(2 c_A \alpha + 2 c_B \beta\right) - E \left(2 c_A + 2 c_B S\right) \\
&= 2 \left( c_A \alpha + c_B \beta - E c_A - E S c_B\right) \\
&= 2 \left( c_A (\alpha - E) + c_B (\beta - SE ) \right)
\end{aligned}
$$

Analog folgt für die Ableitung nach $c_B$:

$$
\frac{\partial E}{\partial c_B} = 2 \left( c_B (\alpha - E) + c_A (\beta - SE ) \right)
$$

Es gilt $ \frac{\partial E}{\partial c_A} = \frac{\partial E}{\partial c_B} = 0$:

$$
\begin{aligned}
2 \left( c_A (\alpha - E) + c_B (\beta - SE ) \right) &= 0 \\
2 \left( c_B (\alpha - E) + c_A (\beta - SE ) \right) &= 0 \\
\end{aligned}
$$

Wir sortieren um:

$$
\begin{aligned}
\left( c_A (\alpha - E) + c_B (\beta - SE ) \right) &= 0 \\
\left( c_A (\beta - SE ) + c_B (\alpha - E) \right) &= 0 \\
\end{aligned}
$$


Dieses Gleichungssystem können wir in Matrixform bringen:

$$
\begin{bmatrix}
\alpha - E & \beta - SE \\
\beta - SE & \alpha - E \\
\end{bmatrix}
\begin{pmatrix} c_A \\ c_B \end{pmatrix}
=
\begin{pmatrix} 0 \\ 0 \end{pmatrix}
$$

Dieses Gleichungssystem ist erfüllt, wenn die Determinante der Matrix verschwindet:

$$
\begin{aligned}
\begin{vmatrix}
\alpha - E & \beta - SE \\
\beta - SE & \alpha - E \\
\end{vmatrix} &= 0 \\
\left(\alpha - E\right)^2 - \left(\beta - SE\right)^2 &= 0 \\
\alpha^2 - 2 \alpha E + E^2 - \beta^2 + 2 \beta SE - E^2 S^2 &= 0
\end{aligned}
$$

Wodurch wir die Energie berechnen können:

$$
\begin{aligned}
E^2 - 2 \alpha E + 2 \beta SE - E^2 S^2 &= - \left( \alpha^2 - \beta^2 \right) \\
E^2 (1 - S^2) + E \left( - 2 \alpha + 2 \beta S \right) &=  - \left( \alpha^2 - \beta^2 \right) \\
E^2 + E \frac{- 2 \alpha + 2 \beta S}{1 - S^2} &= -\frac{\alpha^2 - \beta^2}{1-S^2} \\
\end{aligned}
$$

Dies können wir mit der quadratischen Lösungsformel lösen:

$$
\begin{aligned}
E_{1,2} &= \frac{\alpha - \beta S}{1 - S^2} \pm
\sqrt{
\frac{\left(\alpha - \beta S\right)^2}{\left(1 - S^2\right)^2} - \frac{\alpha^2 - \beta^2}{1-S^2}
}\\
&= \frac{\alpha - \beta S}{1 - S^2} \pm
\sqrt{
\frac{\left(\alpha - \beta S\right)^2}{\left(1 - S^2\right)^2} 
- \frac{\left(\alpha^2 - \beta^2\right) \left(1-S^2\right)}{\left(1-S^2\right)^2}
}\\
&= \frac{\alpha - \beta S}{1 - S^2} \pm
\sqrt{
\frac{\left(\alpha - \beta S\right)^2 
- \alpha^2 + \beta^2
+ S^2 \alpha^2 - S^2 \beta^2}{\left(1 - S^2\right)^2} 
}\\
&= \frac{\alpha - \beta S}{1 - S^2} \pm
\sqrt{
\frac{- 2 \alpha \beta S + \beta^2 + S^2 \alpha^2}{\left(1 - S^2\right)^2} 
}\\
&= \frac{\alpha - \beta S}{1 - S^2} \pm
\sqrt{
\frac{\left(\beta - S \alpha \right)^2}{\left(1 - S^2\right)^2} 
}\\
\end{aligned}
$$

Dies ergibt folgende zwei Lösungen, mit dem Trick $\left(1 - S^2 \right) = \left(1 + S \right)\left(1 - S \right)$:

$$
\begin{aligned}
E_1 &= \frac{\left(\alpha - \beta S\right) +  \left(\beta - S \alpha \right)}{1 - S^2} 
= \frac{\alpha \left(1 - S\right) +  \beta \left(1 - S \right)}{\left(1 + S \right)\left(1 - S \right)}
= \frac{\alpha + \beta}{1 + S}
\\
E_2 &= \frac{\alpha - \beta}{1 - S}
\end{aligned}
$$

Hierbei hat $E_1 = \varepsilon_1$, also die Energie des ersten Molekülorbitals, die geringere Energie und wird zuerst besetzt.

Setzt man diese Eigenwerte in die obige Matrixgleichung ein, lassen sich die Eigenvektoren bestimmen. Diese sind:

$$
\begin{aligned}
E_1 &\rightarrow c_A = + \frac{1}{\sqrt{2}},~ c_B = + \frac{1}{\sqrt{2}} \\
E_2 &\rightarrow c_A = + \frac{1}{\sqrt{2}},~ c_B = - \frac{1}{\sqrt{2}} \\
\end{aligned}
$$

Diese interpretiert man als bindende (konstruktive) und anti-bindende (destruktive) Interferenz.
