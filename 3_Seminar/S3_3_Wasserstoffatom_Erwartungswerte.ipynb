## Erwartungswerte

### Der Erwartungswert der Wahrscheinlichkeit

**Beispiel:** 1s-Orbital im Bereich $a_0 / 2 < r < 3 a_0 / 4 $

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\begin{aligned}
\psi_{100} &= R_{100}(r) Y_{00}(\theta, \varphi)
= \frac{a_0^{-3}}{{\pi}} e^{-r/a_0}\\
\braket{A}_{a_0/2}^{3a_0/4} &= 
\frac{a_0^{-3}}{{\pi}}
\int_0^{2\pi} d\varphi \int_0^\pi \sin(\theta) d\theta
\int_{a_0/2}^{3a_0/4} r^2 e^{-2r/a_0} dr \\
\braket{A}_{a_0/2}^{3a_0/4} &= 
4a_0^{-3}
\int_{a_0/2}^{3a_0/4} r^2 e^{-2r/a_0} dr \\
\vdots \\
\braket{A}_{a_0/2}^{3a_0/4} &= 0.1109~\checkmark
\end{aligned}
$$

\begin{aligned}

\end{aligned}
$$

### Der Erwartungswert der Energie

**Beispiel**: 1s-Orbital
$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\begin{aligned}
\braket{E_{100}} &= \braket{\psi_{100}(r, \theta, \varphi) \left| \hat{H} \right| \psi_{100}(r, \theta, \varphi)} \\
&= \braket{\psi_{100}(r, \theta, \varphi) | \frac{-\Delta}{2} - \frac{1}{|\hat{r}|} | \psi_{100}(r, \theta, \varphi)} \\
&= \int \psi_{100}(r, \theta, \varphi) \frac{-1}{2} \left( 
\frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} \right)
+ \frac{1}{r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial}{\partial \theta} \right) 
+ \frac{1}{r^2 \sin^2(\theta)} \frac{\partial^2 }{\partial \varphi^2} 
- \frac{1}{|\hat{r}|} \right) 
\psi_{100}(r, \theta, \varphi) d\tau
\end{aligned}
$$

Schauen wir uns die kinetische Energie zuerst an:

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\begin{aligned}
\Delta \psi_{100}(r, \theta, \varphi) &= 
\left(\frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} \right)
\underbrace{
+ \frac{1}{r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial}{\partial \theta} \right) 
+ \frac{1}{r^2 \sin^2(\theta)} \frac{\partial^2 }{\partial \varphi^2}}_{=0}
\right)
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} \\
&= \frac{a_0^{-3/2}}{\sqrt{\pi}} \frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial e^{-r/a_0}}{\partial r} \right)  \\
&= \vdots \\
&= \frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} \left( \frac{1}{a_0^2} - \frac{2}{r a_0} \right)
\end{aligned} 
$$

Jetzt müssen wir folgende Funktion integrieren (mit SI-Einheiten):

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\begin{aligned}
\braket{E_{100}} &= \frac{a_0^{-3}}{\pi}  
\int \left( 
- \frac{\hbar^2 e^{-r/a_0} \Delta e^{-r/a_0}}{2m_e} 
- e^{-r/a_0} \cdot \frac{\hbar^2}{m_e a_0|\hat{r}|} e^{-r/a_0} \right)
d\tau \\
&= \frac{a_0^{-3} \hbar^2}{\pi m_e}  \int \left( 
e^{-2r/a_0} \left( 
\frac{-1}{2 a_0^2} + \frac{1}{r a_0} \right)
- \frac{e^{-2r/a_0}}{r a_0} \right)   d \tau \\
&= \frac{a_0^{-3} \hbar^2}{\pi m_e}  \int e^{-2r/a_0} \left( 
- \frac{1}{2 a_0^2} + \frac{1}{r a_0} - \frac{1}{r a_0} \right) d \tau \\
&= \vdots \\
&= -0.5~E_H = -13.6~eV~~\checkmark
\end{aligned}
$$

### Der Erwartungswert des Abstands

Der Ortsoperator der Wellenfunktion agiert folgendermaßen:

$$
\hat{x}_i \Psi(x_i) = x_i \cdot \Psi(x_i)
$$

Wir können also einen Abstandsoperator als den Betrag des Ortsoperators definieren und in Kugelkoordinaten darstellen:

$$
\begin{aligned}
|\hat{x}_i| \Psi(x_i) &= |x_i| \cdot \Psi(x_i) \\
\hat{r} \Psi(r, \theta, \phi) &= |\mathbf{r}| \cdot \Psi(r, \theta, \phi) \\
&= \sqrt{
\left[|r| \sin(\theta)\cos(\phi)\right]^2 
+ \left[|r| \sin(\theta)\sin(\phi)\right]^2 
+ \left[|r| \cos(\theta)\right]^2} \cdot \Psi(r, \theta, \phi) \\
&= \sqrt{|r|^2 \cdot \underbrace{
\left[ 
\sin(\theta)^2 \cdot 
\left( \cos(\phi)^2 + \sin(\phi)^2 \right) 
+ \cos(\theta)^2
\right]}_{=1}} \cdot \Psi(r, \theta, \phi) \\
&= r \cdot \Psi(r, \theta, \phi)
\end{aligned}
$$

Der Erwartungswert des Abstands in Kugelkoordinaten ist also:

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\braket{r} = \braket{\Psi|\hat{r}|\Psi} = \braket{\Psi|r|\Psi} = \int r \cdot |\Psi|^2 d\tau
$$

Für das 1s-Orbital des Wasserstoffatoms erhält man:

$$
\newcommand{\ket}[1]{\left|{#1}\right\rangle}
\newcommand{\bra}[1]{\left\langle{#1}\right|}
\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}
\begin{aligned}
\braket{r} &= \frac{4}{a_0^3} \int_0^\infty r^3 e^{-2r/a_0} dr \\
&= \frac{4}{a_0^3} \cdot \frac{6}{\left( \frac{2}{a_0} \right)^4} =\frac{3}{2} a_0 \\
&\approx 1.5 ~ \text{Angström}
\end{aligned}
$$

Als Vergleich, der van-der-Waals-Radius von Wasserstoff liegt zwischen 1.2 bis 2 Angström.
