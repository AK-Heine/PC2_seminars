# Seminar 3: Das Wasserstoffatom

## Lösen der Schrödinger-Gleichung

Die Schrödinger-Gleichung für das Wasserstoffatom lautet:

$$
\left( \underbrace{\frac{-1}{2} \Delta}_{=\hat{T}_{kin}}  - \underbrace{\frac{1}{| \mathbf{\hat{r}}|}}_{=\hat{V}_{Coulomb}} \right) \Psi(\mathbf{r}) = E\Psi(\mathbf{r})
$$

Hier ist $\mathbf{r} = x \hat{i} + y \hat{j} + z \hat{k} = \left(x~y~z\right)^T$ ein Vektor in kartesischen Koordinaten. Da es sich aber um ein zentrosymmetrisches Problem handelt, bietet sich die Transformation zu Kugelkoordinaten an:

$$
\begin{aligned}
x &\rightarrow |r| \cdot \sin(\theta) \cos(\varphi)  ~~ &r \in \left[0, \infty\right] \\
y &\rightarrow |r| \cdot \sin(\theta) \sin(\varphi)  ~~ &\theta \in \left[0, \pi\right] \\
z &\rightarrow |r| \cdot \cos(\theta)             ~~ &\varphi \in \left[0, 2\pi\right]
\end{aligned}
$$

Das bedeutet, die Wellenfunktion transformiert nach $\Psi(\mathbf{r}) \rightarrow \Psi(r, \theta, \varphi)$. Der Coulomb-Operator $\hat{V}$ ist winkelunabhängig und transformiert einfach zu $1/r$. Der Laplace-Operator für die kinetische Energie sieht in Kugelkoordinaten allerdings anders aus:

$$
\Delta = \frac{1}{r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} \right)
+ \frac{1}{r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial}{\partial \theta} \right) 
+ \frac{1}{r^2 \sin^2(\theta)} \frac{\partial^2 }{\partial \varphi^2}
$$

Dieser Operator suggeriert bereits, dass wir die radiale und winkelabhängige Bewegung separieren können:

$$
\Psi(r, \theta, \varphi) = R(r) \cdot Y(\theta, \varphi)
$$

Setzen wir dies in die SG ein, folgt:

$$
\left( \frac{-1}{2r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial}{\partial r} \right)
+ \frac{-1}{2r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial}{\partial \theta} \right) 
+ \frac{-1}{2r^2 \sin^2(\theta)} \frac{\partial^2 }{\partial \varphi^2} 
+ \hat{V} \right) R(r) \cdot Y(\theta, \varphi) = E R(r) \cdot Y(\theta, \varphi)
$$

Das sieht viel komplizierter aus, als es eigentlich ist. Die winkelabhängigen Operatoren kommutieren mit $R(r)$, die winkelunabhängigen Operatoren kommutieren mit $Y(\theta, \varphi)$:

$$
\begin{aligned}
&Y \cdot \frac{-1}{2r^2} \frac{\partial}{\partial r} \left(r^2 \frac{\partial R}{\partial r} \right) \\
&+ R \cdot \frac{-1}{2r^2 \sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial Y}{\partial \theta} \right) \\
&+ R \cdot \frac{-1}{2r^2 \sin^2(\theta)} \frac{\partial^2 Y}{\partial \varphi^2} \\
&+ Y \cdot \hat{V} R  \\
&= R Y \cdot E 
\end{aligned}
$$

Wir multiplizieren mit $- 2\cdot r^2$ und dividieren durch $RY$:

$$
\begin{aligned}
\underbrace{
R^{-1}\left[ \frac{\partial}{\partial r} \left(r^2 \frac{\partial R}{\partial r} \right) 
-2\cdot r^2 \left( \hat{V} R - E \right) \right]
}_{\text{radialabhängig}}
+ \underbrace{
Y^{-1} \left[ \cdot \frac{1}{\sin(\theta)} \frac{\partial}{\partial \theta} \left(\sin(\theta) \frac{\partial Y}{\partial \theta} \right) 
+\cdot \frac{1}{\sin^2(\theta)} \frac{\partial^2 Y}{\partial \varphi^2}\right]
}_{\text{winkelabhängig}} = 0
\end{aligned}
$$



## Wie sehen die Lösungen aus?

Beide Seiten müssen gleich groß und konstant sein. Die winkelabhängige Gleichung ist dabei eine spezielle Differentialgleichung: Die Laplace-Gleichung aus der klassischen Elektrodynamik. Deren Lösungen sind die sogenannten Legendre-Funktionen (Kugelflächenfunktionen, "spherical harmonics"). Mit dieser Information lassen sich auch die Radialfunktionen exakt bestimmen.

Die Kugelflächenfunktionen haben folgende Form:

$$
Y_{l, m}(\theta, \varphi) = \underbrace{\epsilon \sqrt{\frac{(2l+1)}{4\pi} \frac{(l - |m|)!}{(l + |m|)!}}}_{\text{Normierung}} ~\cdot~ \underbrace{e^{im\varphi}}_{\text{komplexe Phase}}~\cdot~
\underbrace{P_l^m(cos(\theta))}_{\text{Legendre-Funktionen}}
~~\text{mit}~\epsilon=\begin{cases}(-1)^m~\forall~m\geq 0 \\ 1~\forall~m < 0 \end{cases}
$$

Die Lösungen der Radialfunktion haben folgende Form ($a_0$ ist der Bohr-Radius):

$$
R_{n,l}(r) = \underbrace{\sqrt{\frac{2}{na_0}^3 \frac{(n-l-1)!}{2n[(n+l)!]^3}}}_{\text{Normierung}}~\cdot~
\underbrace{\left(\frac{2r}{na_0}\right)^l \left[L_{n-l-1}^{2l+1}({2r}/{na_0})\right]}_{\text{Laguerre-Polynome}}~\cdot~
\underbrace{e^{-r/na_0}}_{\text{exponentieller Abfall}}
$$

Mit folgenden Quantenzahlen:

$$
\begin{aligned}
n &= 1,2,3... & \rightarrow \text{Hauptquantenzahl} \\
l &= 0,1,2... <n & \rightarrow \text{Nebenquantenzahl} \\
m &= -l, (-l+1), ..., (l-1), +l & \rightarrow \text{Magnetquantenzahl} \\
\end{aligned}
$$

Die Energieniveaus hängen nur von der Hauptquantenzahl $n$ ab:

$$
E_{n} = \left[ \frac{-m_e e^4}{8 \varepsilon_0^2 h^2} \right] \frac{1}{n^2}
$$

Die Lösungen der Schrödinger-Gleichung für das Wasserstoffatom sind somit nicht besonders schön. Stark vereinfacht können wir sie uns folgendermaßen vorstellen:

$$
\Psi_{n, l, m} = \text{Abstandsabhängigkeit} \times \text{Raumausrichtung} \times \text{Quantisierung}
$$

Als praktisches Beispiel:
Der Realteil der Kugelflächenfunktion für $l=1$ und $m_l=0$ ($p$-Orbital) sieht so aus:

$$
Y_{l=1,m=0} = \sqrt{\frac{3}{4\pi}} \cos(\theta)
$$

Die Radialfunktion für das dazugehörige $n=2$ und $l=1$ sieht so aus:

$$
R_{n=2, l=1} = \frac{1}{\sqrt{24}} a_0^{-3/2} \frac{r}{a} e^{-r/2a_0}
$$

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.special import sph_harm, genlaguerre, factorial
from matplotlib import cm, colors
plt.xkcd()
plt.style.use("seaborn-ticks")
plt.rcParams["legend.handlelength"] = 0.8
plt.rcParams["legend.framealpha"] = 0.8
font_size = 12
plt.rcParams.update({"font.size": font_size})

def R_21(r):
    return 1/np.sqrt(24) * 1**(-3/2) * r * np.exp(-r/2)

def Y_10(theta):
    return  np.cos(theta) * np.sqrt(3/(4*np.pi))

theta = np.linspace(0, 2, 100) * np.pi
r = np.linspace(0, 20, 100)

fig = plt.figure(figsize=(15, 5))
ax = plt.subplot(1, 3, 1)
ax.plot(r, R_21(r))
ax.title.set_text("Radial part")
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel("r")
ax.set_ylabel("$R_{21}(r)$")

ax = plt.subplot(1, 3, 2, projection='polar')
ax.plot(theta+(r<0)*np.pi, np.abs(Y_10(theta)))
ax.set_rticks([])  # less radial ticks
ax.set_rlabel_position(-22.5)  # get radial labels away from plotted line
ax.grid(True)
ax.title.set_text("Angular part")

Z   = 1  # Anzahl der Elementarladungen
a_0 = 1. # Bohrscher Atomradius

def R(n,l, r):
    """Radialfunktion"""
    Z_n = 2*Z/(n*a_0)
    root = np.sqrt(Z_n**3 * factorial(n-l-1)/(2*n*factorial(n+l)))
    rho = Z_n * r
    L = np.polyval(genlaguerre(n-l-1, 2*l+1), rho)
    return root * np.exp(-0.5*rho) * rho**l * L

def psi(n,l,m, r,theta,phi):
    """de.wikipedia.org/wiki/Wasserstoffatom#Mathematische_Details"""
    R_nl = R(n,l, r)
    Y_lm = sph_harm(m,l,phi,theta) 
    return R_nl * Y_lm

def plot(n,l,m, x,z, r,theta,phi, ax):
    ax.title.set_text(r"$|R(r) \cdot Y(\theta, \varphi )|^2$")
    wf = psi(n,l,m, r,theta,phi)
    Psi = np.abs(wf * wf.conj())
    ax.pcolormesh(x, z, Psi, cmap=plt.get_cmap("seismic_r"), 
                  norm = colors.Normalize(vmin=-0.001,vmax=0.001))
    plt.gca().set_aspect("equal") 
    ax.set_xlim(x.min(),x.max())    
    ax.set_ylim(z.min(),z.max())
    
ax = plt.subplot(1, 3, 3)
ax.set_xticks([])
ax.set_yticks([])

n = 2
l = 1
m = 1

plot_range = (4*n + 4*l) * a_0
x_1d  = np.linspace(-plot_range,plot_range,521)
y     = 0
z_1d  = np.linspace(-plot_range,plot_range,521)
x,z   = np.meshgrid(x_1d,z_1d)
r     = np.sqrt(x**2 + y**2 + z**2)
theta = np.arctan2(np.sqrt(x**2+y**2), z )
phi   = np.arctan2(y, x)
plot(n,l,m, x,z, r,theta,phi,ax)

plt.show()
```

### Lyman-Serie
Die Lyman-Serie ist eine spezielle spektroskopische Serie des Wasserstoffatoms, wo Übergänge von angeregten Zuständen $n'$ zum Grundzustand $n=1$ beobachtet werden.

**Ansatz:** Rydberg-Formel:

$$
\frac{1}{\lambda} = R_\text{H} \left( \frac{1}{n^2} - \frac{1}{n'^2} \right) 
$$

Mit der Rydberg-Konstante $R_\text{H} \approx ~1.0968\cdot 10^7~\text{m}^{-1}~\approx~\frac{13.6~\text{eV}}{hc}$ und für die Übergangsenergie $\lambda = 9.38\cdot 10^{-8}~m$ soll $n'$ bestimmt werden:

$$
\begin{aligned} 
\frac{1}{n^2} - \frac{1}{R_\text{H} \cdot \lambda}  &=  \frac{1}{n'^2} \\
\vdots \\
n' &= 5.977 \approx 6
\end{aligned}
$$

### Ionisierungsenergie

Die Ionisierungsenergie $IE$ ist die Energie, die man aufbringen muss, um ein Elektron von einem bestimmten Energieniveau ins Vakuum zu entfernen. Der Grundzustand des Wasserstoffatoms hat ein Elektron im Zustand $n=1$. Wir müssen also das Elektron von $n=1 \rightarrow n'=\infty$ bringen.

$$
\begin{aligned}
\Delta E &= \left[ \frac{-m_e e^4}{8 \varepsilon_0^2 h^2} \right] \left( \frac{1}{n^2} - \frac{1}{n'^2} \right) \\
\Delta E &= \lim_{n'\rightarrow \infty, n=1} \left[ \frac{-m_e e^4}{8 \varepsilon_0^2 h^2} \right] \left( \frac{1}{n^2} - \frac{1}{n'^2} \right)\\
&= 1 R_\text{H} \\
&= 13.6058~\text{eV}\\
&= 0.5~E_H \\
&= 13.12751~kJ/mol \\
&= 313.755~kcal/mol 
\end{aligned}
$$
